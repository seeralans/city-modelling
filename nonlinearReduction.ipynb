```python
# Import all necessary packages
import math
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
from matplotlib import pyplot as plt
import glob
from sklearn.decomposition import PCA
import sklearn.manifold as sman
from pandas.tools.plotting import scatter_matrix

import matplotlib
matplotlib.style.use('ggplot')

%matplotlib inline
```

```python
data = pd.read_csv("./Data/Data/Reduced_Census/Census/ks101ew")
```

```python
frame = pd.DataFrame()
dataList = []
fileNames = glob.glob('./Data/Reduced_Census/Census/ks*')
for file in fileNames:
    data = pd.read_csv(file, index_col='GeographyCode')
    dataList.append(data)
    

frame = pd.concat(dataList, axis=1)

```

```python
pca = PCA(n_components=5)
pca.fit(frame)
frame.isnull().values.any()
X = pca.fit_transform(frame)
plt.scatter(X[:,1], X[:,0])
```

```python
XX = pd.DataFrame(X)
scatter_matrix(XX, alpha=0.2, figsize=(8, 8), diagonal="kde")
plt.show()
```

```python
isomap = sman.Isomap(n_neighbors=5, n_components=5)
XT = isomap.fit_transform(frame)

```

```python
plt.scatter(XT[:,0], XT[:,1])
```

```python
XXT = pd.DataFrame(XT)
scatter_matrix(XXT, alpha=0.2, figsize=(8, 8), diagonal="kde")
plt.show()
```

```python
linearEmbed = sman.LocallyLinearEmbedding(5, 5)
XT = linearEmbed.fit_transform(frame)
```

```python
XXT = pd.DataFrame(XT)
scatter_matrix(XXT, alpha=0.2, figsize=(8, 8), diagonal="kde")
plt.show()
```

```python
spect = sman.SpectralEmbedding(n_components=5)
XT = spect.fit_transform(frame)
```

```python
XXT = pd.DataFrame(XT)
scatter_matrix(XXT, alpha=0.2, figsize=(8, 8), diagonal="kde")
plt.show()
```

```python
mds = sman.MDS(n_components=5)
XT = mds.fit_transform(frame)
XXT = pd.DataFrame(XT)
scatter_matrix(XXT, alpha=0.2, figsize=(8, 8), diagonal="kde")
plt.show()
```

```python
tsne = sman.TSNE(n_components=5)
XT = tsne.fit_transform(frame)
XXT = pd.DataFrame(XT)
scatter_matrix(XXT, alpha=0.2, figsize=(8, 8), diagonal="kde")
plt.show()
```
